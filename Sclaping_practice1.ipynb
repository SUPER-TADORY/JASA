```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
df=pd.read_csv("japan.csv")
df
```

```python
cols_to_fill=list(df.columns)
cols_to_fill
```

```python
def func(i,stock_code):
    driver = webdriver.Chrome()
    driver.get(f'https://finance.yahoo.co.jp/quote/{int(stock_code)}.T')
    
    soup = BeautifulSoup(driver.page_source, features="html.parser")
    
    name=soup.findAll("h1",{"class":"_6uDhA-ZV"})[0].text
    l=soup.findAll("span",{"class":"_3rXWJKZF _11kV6f2G"})
    
    df.loc[i,cols_to_fill[1]]=name
    
    df.loc[i,cols_to_fill[2]]=l[1].text
    df.loc[i,cols_to_fill[3]]=l[2].text
    df.loc[i,cols_to_fill[4]]=l[3].text
    df.loc[i,cols_to_fill[5]]=l[0].text
    df.loc[i,cols_to_fill[6]]=l[5].text
    
    driver.close()
    
    print(name,"：正常終了")
    
```

```python
import time

for i,row in df.iterrows():
    func(i,row["Stock code"])
    time.sleep(2)
```
