```python
import requests
from bs4 import BeautifulSoup
import time

def func(i,url):
    time.sleep(2) 
    
    option = Options()                          # オプションを用意
    option.add_argument('--headless')           # ヘッドレスモードの設定を付与
    driver = webdriver.Chrome()
    driver.get(url)

    soup = BeautifulSoup(driver.page_source, features="html.parser")
    profile=soup.find("span",{"class":"company_summary"}).text
    driver.close()
           
    df1.loc[i,"profile"]=profile
    
    print(i,"：正常終了")
```

```python
mic_df=df1[190:213]

for i,row in mic_df[mic_df["D&B URL"].notna()].iterrows():
    url=row["D&B URL"]
    try:
        func(i,url)
    except:
        print(i)
```

```python
for i,row in df1[df1["profile"].notna()].iterrows():
    tmp=row["profile"]
    
    earn_pattern1 = r"([0-9]+)\,([0-9]+)\.([0-9]+) million"
    earn_pattern2 = r"([0-9]+)\.([0-9]+) million"
    earn_pattern3 = r"([0-9]+)\,([0-9]+) in sales"
    earn_pattern4 = r"([0-9]+)\.([0-9]+) billion"
    
    employ_pattern2 = r"([0-9]+) ([a-z]+) employee"
    employ_pattern1 = r"([0-9]+),([0-9]+) ([a-z]+) employee"
    employ_pattern4 = r"([0-9]+) employee"
    employ_pattern3 = r"([0-9]+),([0-9]+) employee"
    
    m1 = re.search(earn_pattern1, tmp)
    m2 = re.search(earn_pattern2, tmp)
    m3 = re.search(earn_pattern3, tmp)
    m4 = re.search(earn_pattern4, tmp)
    
    n1 = re.search(employ_pattern1, tmp)
    n2 = re.search(employ_pattern2, tmp)
    n3 = re.search(employ_pattern3, tmp)
    n4 = re.search(employ_pattern4, tmp)
    
    n=n1 if n1 else n2 if n2 else n3 if n3 else n4
    m=m1 if m1 else m2 if m2 else m3 if m3 else m4
    
    try:
        print(i,"earn : ",m[0],"employ : ",n[0])
        #print(i,"employ : ",n[0].split()[0])
        #print(i,"earn : ",m[0])
    except:
        if m:
            print(i,"earn : ",m[0],row["D&B URL"])
        elif n:
            print(i,"employ : ",n[0],row["D&B URL"])
        else:
            print(i,row["D&B URL"])
```

```python
import openpyxl
openpyxl.Workbook()
wb=openpyxl.load_workbook(data_path)
sheet = wb["AU Main Long List"]
sheet
```

```python
for i,row in df1[df1["profile"].notna()].iterrows():
    print(i,"##############################")
    tmp=row["profile"]
    
    earn_pattern1 = r"([0-9]+)\,([0-9]+)\.([0-9]+) million"
    earn_pattern2 = r"([0-9]+)\.([0-9]+) million"
    earn_pattern3 = r"([0-9]+)\,([0-9]+) in sales"
    earn_pattern4 = r"([0-9]+)\.([0-9]+) billion"
    
    employ_pattern2 = r"([0-9]+) ([a-z]+) employee"
    employ_pattern1 = r"([0-9]+),([0-9]+) ([a-z]+) employee"
    employ_pattern4 = r"([0-9]+) employee"
    employ_pattern3 = r"([0-9]+),([0-9]+) employee"
    
    m1 = re.search(earn_pattern1, tmp)
    m2 = re.search(earn_pattern2, tmp)
    m3 = re.search(earn_pattern3, tmp)
    m4 = re.search(earn_pattern4, tmp)
    
    n1 = re.search(employ_pattern1, tmp)
    n2 = re.search(employ_pattern2, tmp)
    n3 = re.search(employ_pattern3, tmp)
    n4 = re.search(employ_pattern4, tmp)
    
    n=n1 if n1 else n2 if n2 else n3 if n3 else n4
    m=m1 if m1 else m2 if m2 else m3 if m3 else m4
    
    if n:
        out1=n[0].split()[0]
        sheet.cell(column=14,row=i+2).value = out1
        print("1",out1)
    
    if m:
        if "million" in m[0]:
            out2=" ".join(m[0].split()[:2])
            sheet.cell(column=22,row=i+2).value = out2
        
        elif "billion" in m[0]:
            out2=" ".join(m[0].split()[:2])
            sheet.cell(column=22,row=i+2).value = out2
         
        elif "sale" in m[0]:
            out2=m[0].split()[0]
            sheet.cell(column=22,row=i+2).value = out2
        print("2",out2)
        
    if m:
        if "million" in m[0]:
            out2=m[0].split()[0]
            sheet.cell(column=22,row=i+2).value = out2

        elif "billion" in m[0]:
            out2=int(float(m[0].split()[0]))*1000
            sheet.cell(column=22,row=i+2).value = out2

        elif "sale" in m[0]:
            a=int(float("".join(m[0].split()[0].split(","))))
            out2="{:.2f}".format(a/1000000)
            sheet.cell(column=22,row=i+2).value = out2
```
