```python
import pandas as pd
import numpy as np
import sys
import os

from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
df=pd.read_csv("japan.csv")
df
```

```python
cols_to_fill=list(df.columns)
cols_to_fill
```

```python
def func(i,stock_code):
    driver1 = webdriver.Chrome()
    driver1.get(f'https://jp.investing.com/search/?q={stock_code}')

    soup1 = BeautifulSoup(driver1.page_source, features="html.parser")
    ext1=soup1.findAll("a",{"class":"js-inner-all-results-quote-item row"})[0]
    link=ext1.get("href")
    name=ext1.find("span",{"class":"third"}).text
    driver1.close()
    
    driver2 = webdriver.Chrome()
    driver2.get(f'https://jp.investing.com{link}')
    soup2 = BeautifulSoup(driver2.page_source, features="html.parser")
    l=soup2.findAll("span",{"class":"key-info_dd-numeric__2cYjc"})
           
    df.loc[i,cols_to_fill[1]]=name
    
    df.loc[i,cols_to_fill[2]]=l[4].text
    df.loc[i,cols_to_fill[3]]=l[2].text
    df.loc[i,cols_to_fill[4]]=l[1].text
    df.loc[i,cols_to_fill[5]]=l[0].text
    df.loc[i,cols_to_fill[6]]=l[8].text
    
    driver2.close()
    
    print(name,l[4].text)
    
    print(name,"：正常終了")
```

```python
import time

for i,row in df[80:].iterrows():
    try:
        func(i,row["Stock code"])
        time.sleep(2)
    except:
        print(i,row["Stock code"])
```
